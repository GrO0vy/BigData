# **에어비엔비 데이터를 활용한 미국 주요 관광지 숙박업소 정보 분석**

**1. 원본 데이터 확인<br>**
  Kaggle 사이트에서 가져온 미국 에어비엔비 숙박업소 정보를 확인한다.

```python
import pandas as pd
us_data = pd.read_csv("/content/AB_US.csv")
print(us_data)
```

**2. 데이터 가공<br>**
    숙박업소의 개수 , 가격 , 리뷰 수를 주(state) 단위로 분석할 것이기 때문에 <br>
    각 도시의 숙박업소 개수를 얻어내고 주 단위로 묶기 위해 데이터를 가공

```python
us_city = us_data.drop_duplicates(['city'])
print(us_city)
count = 0
for i in us_city['city']:
  us_data[i] = 0
  for j in us_data['city']:
    if j == i:
      count+=1
  us_data[i]=count
  count=0

print(us_city)
print("===============================================================================================================\n")
print(us_data)
```

```python
us_data['state']=0
index_number=0
for i in us_data['city']:
  if i == "Asheville":
    us_data['state'].iloc[index_number] = "North Carolina"
  if i == "Austin":
    us_data['state'].iloc[index_number] = "Texas"
  if i == "Boston" or i=="Cambridge" or i=="Salem":
    us_data['state'].iloc[index_number] = "Massachusetts"
  if i == "Broward County":
    us_data['state'].iloc[index_number] = "Florida"
  if i == "Chicago":
    us_data['state'].iloc[index_number] = "Illinois"
  if i == "Clark County":
    us_data['state'].iloc[index_number] = "Nevada"
  if i == "Columbus":
    us_data['state'].iloc[index_number] = "Ohio"
  if i == "Denver":
    us_data['state'].iloc[index_number] = "Colorado"
  if i == "Jersey City":
    us_data['state'].iloc[index_number] = "New Jersey"
  if i == "Los Angeles" or i =="Oakland" or i =="Pacific Grove"or i =="San Clara Country"or i =="San Diego"or i =="San Francisco" or i=="San Mateo County" or i== "Santa Cruz County":
    us_data['state'].iloc[index_number] = "California"
  if i == "Nashville":
    us_data['state'].iloc[index_number] = "Tennessee"
  if i == "New Orleans":
    us_data['state'].iloc[index_number] = "Louisiana"
  if i == "New York City":
    us_data['state'].iloc[index_number] = "New York"
  if i == "Portland":
    us_data['state'].iloc[index_number] = "Oregon"
  if i == "Seattle" or i =="Washington D.C.":
    us_data['state'].iloc[index_number] = "Washington"
  if i == "Twin Cities MSA":
    us_data['state'].iloc[index_number] = "Minnesota"
  if i == "Hawaii":
    us_data['state'].iloc[index_number] = "Hawaii"
  if i == "Rhode Island":
    us_data['state'].iloc[index_number] = "Rhode Island"
  index_number +=1
print(us_data)
```

```python
new_us_data = pd.DataFrame()
new_us_data['country']=0
new_us_data['count']=0

city_counts=[]
city_list=[]
for i in us_city['city']:
  city_list.append(i)
  city_counts.append(us_data[i].iloc[1])

new_data = {
    'country':city_list,
    'count':city_counts
}

new_us_data = pd.DataFrame(new_data)

map_data = pd.DataFrame()
map_data['country']=0
map_data['count']=0

new_city_list=["North Carolina","Texas","Massachusetts","Florida","Hawaii","Illinois","Nevada","Ohio","Colorado","New Jersey",
              "California","Tennessee","Louisiana","New York","Oregon","Rhode Island","Washington","Minnesota"]
new_city_counts=[
    us_data['Asheville'].iloc[1],us_data['Austin'].iloc[1],us_data['Boston'].iloc[1]+us_data['Cambridge'].iloc[1]+us_data['Salem'].iloc[1],
     us_data['Broward County'].iloc[1],us_data['Hawaii'].iloc[1] ,us_data['Chicago'].iloc[1], us_data['Clark County'].iloc[1],us_data['Columbus'].iloc[1],
     us_data['Denver'].iloc[1], us_data['Jersey City'].iloc[1],
     us_data['Los Angeles'].iloc[1]+us_data['Oakland'].iloc[1]+us_data['Pacific Grove'].iloc[1]+us_data['San Diego'].iloc[1]+us_data['San Francisco'].iloc[1]+
     us_data['San Mateo County'].iloc[1]+us_data['Santa Cruz County'].iloc[1]+us_data['San Clara Country'].iloc[1],
     us_data['Nashville'].iloc[1],us_data['New Orleans'].iloc[1],us_data['New York City'].iloc[1],us_data['Portland'].iloc[1],us_data['Rhode Island'].iloc[1],
     us_data['Seattle'].iloc[1]+us_data['Washington D.C.'].iloc[1],us_data['Twin Cities MSA'].iloc[1]
]


map_list = {
    'country' : new_city_list,
    'count' : new_city_counts
}
map_data = pd.DataFrame(map_list)
print(map_data)
```

```python
import numpy as np
review_count_data = pd.DataFrame()
review_count_data['state']=0
review_count_data['avg_review_per_month']=0
sum=0
count=0
state_list=[]
avg_list=[]
new_state_data = us_data.drop_duplicates(['state'])
print(new_state_data)
for i in new_state_data['state']:
  state_list.append(i)
  for j,k in zip(us_data['state'],us_data['reviews_per_month'].replace(np.nan,'N')):
    if j == i:
      if not k=='N':
        sum+=k
        count+=1 
  avg_list.append(sum/count)
  sum=0
  count=0
print(avg_list)
review_data = {
    'state' : state_list,
    'avg_review_per_month' : avg_list
}
review_count_data = pd.DataFrame(review_data)
print(review_count_data)   
```

```python
us_map = folium.Map(location=[35.300039, -112.455542], zoom_start=4.48)

folium.Choropleth(
  geo_data=us_geo, # 지도 경계
  data = review_count_data, # 표시하려는 데이터
  columns = ['state','avg_review_per_month'], # 열 지정
  fill_color='YlOrRd',
  key_on='feature.properties.name',
).add_to(us_map)

temp=[]
index_number =0
review_max_list=[]
review_min_list=[]
for key, value in locs.items():
  for j,k in zip(us_data['state'] ,us_data['reviews_per_month']):
    if j == review_count_data['state'].iloc[index_number]:
      if not k ==0 :
        temp.append(k)
  review_max_list.append(max(temp))
  review_min_list.append(min(temp))
  folium.map.Marker(
      [value[0], value[1]],
      icon=folium.DivIcon(
          icon_size=(0, 0),
              icon_anchor=(0, 0),
              html='<div\
                      style="\
                          font-size: 0.8rem;\
                          color: black;\
                          background-color:rgba(255, 255, 255, 0);\
                          width:70px;\
                          text-align:center;\
                          margin:0px;\
                      "><b>'
                        +key + '<br>'+str(round(review_count_data['avg_review_per_month'].iloc[index_number],1))+      
              '</b></div>',
      ),
      popup = 'max :' +str(max(temp))+"개" +"\n"+"min: "+str(min(temp))+"개"
  ).add_to(us_map)
  temp=[]
  index_number+=1

us_map
```

```python
price_data = pd.DataFrame()
price_data['state']=0
price_data['avg_price']=0
sum=0
count=0
price_list=[]
print(state_list)
for i in state_list:
  for j,k in zip(us_data['state'],us_data['price'].replace(np.nan,'N')):
    if j == i:
      if not k=='N':
        sum+=k
        count+=1 
  price_list.append(sum/count)
  sum=0
  count=0
print(price_list)
avg_price_data = {
    'state' : state_list,
    'avg_price' : price_list
}
price_data = pd.DataFrame(avg_price_data)
print(price_data)   
```
